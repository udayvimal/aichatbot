```python
from app import build_qa_chain

qa_chain = build_qa_chain()
history = []

while True:
    user_query = input("\nYou: ")
    if user_query.lower() in ["exit", "quit"]:
        print("Chat ended.")
        break

    contexted_query = ""
    for turn in history[-2:]:
        contexted_query += f"User: {turn['user']}\nAI: {turn['bot']}\n"
    contexted_query += f"User: {user_query}"

    result = qa_chain.invoke({"query": contexted_query})
    bot_reply = result["result"]
    print("Bot:", bot_reply)

    history.append({"user": user_query, "bot": bot_reply})
```
